```python
import pandas as pd

rows = []

for tid in raga_ds.track_ids:
    t = raga_ds.track(tid)

    rows.append({
        "track_id": t.track_id,
        "artist": t.artist,
        "raga": t.raga,
        "tradition": t.tradition,
        "audio_path": t.audio_path
    })

df = pd.DataFrame(rows)

df.head()
```

```python
df["raga"].value_counts()
```

```python
selected_ragas = [
    "Tōḍi",
    "Kalyāṇi",
    "Śankarābharaṇaṁ",
    "Karaharapriya",
    "Harikāmbhōji",
    "Māyāmāḷavagauḷa",
    "Kāmavardani",
    "Bhairavi",
    "Ānandabhairavi",
    "Sahānā",
    "Bēgaḍa",
    "Kāṁbhōji",
    "Rītigauḷa",
    "Mōhanaṁ",
    "Hamsadhwani",
    "Hindōḷaṁ",
    "Madhyamāvati",
    "Nāṭakurinji",
    "Kāpi",
    "Dhanyāsi"
]
```

```python
df_20= df[df["raga"].isin(selected_ragas)]
df_20
```

```python
LOCAL_AUDIO_ROOT = "C:/Users/Sragv/MIR Carnatic raga identification/data/raw_audio/Indian Art Music Raga Recognition Dataset (audio)"
```

```python
def rewrite_path(original_path):
    # Find where RagaDataset begins
    idx = original_path.find("RagaDataset")

    if idx == -1:
        raise ValueError(f"'RagaDataset' not found in path: {original_path}")

    relative_part = original_path[idx:]

    # Build new absolute path
    return f"{LOCAL_AUDIO_ROOT}/{relative_part}"
```

```python
df_20["audio_path"] = df_20["audio_path"].apply(rewrite_path)
```
